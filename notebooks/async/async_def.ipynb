# python异步编程 -- 理解协程函数和协程对象
## 什么是协程函数

很简单， 用`async def` 定义的函数就是协程函数。
例如下面定义的func1()就是协程函数

```python
# 1. 定义一个协程函数
async def func1():
    print("Hello, async")
    return "result of async!"
```

## 什么是协程对象

调用一个协程函数并不会立即执行它，而是会返回一个**协程对象**。只有当这个对象被事件循环驱动时（例如通过`await`），函数内的代码才会执行。

```python
# 1. 定义一个协程函数
async def func1():
    print("这行代码在函数被 await 之前不会执行")
    return "Hello, async!"

# 2. 调用协程函数，得到一个协程对象
coro = func1()

# 3. 打印它的类型，确认它是一个协程对象
print(f"协程对象的类型是: {type(coro)}")
print("--- 函数体内的print语句此时并未执行 ---")

# 4. (在Jupyter中) 使用 await 来运行协程并获取结果
result = await coro
print(f"协程执行完毕，结果是: {result}")
```
