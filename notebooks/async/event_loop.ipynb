```python
import src.configs.config
from loguru import logger
import asyncio

async def func1():
   logger.info("step 1")
   await asyncio.sleep(2)
   logger.info("step 2")

async def func2():
   logger.info("step 3")
   await asyncio.sleep(2)
   logger.info("step 4")

def get_event_loop_tasks():
   tasks = [
       func1(),
       func2(),
   ]
   return tasks
```

```python
async def main():
   tasks = get_event_loop_tasks()
   # 在Jupyter中，我们使用 await 来将并发任务交给正在运行的事件循环
   await asyncio.gather(*tasks)

# 在顶层直接 await main 函数
await main()
```
